```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import balanced_accuracy_score
```

```python
df_train = pd.read_csv('/content/train.csv')  # Load training data
df_test = pd.read_csv('/content/test.csv')    # Load test data
```

```python
df_train.head()
df_test.head()
```

```python
X = df_train.drop(['Id', 'Target'], axis=1)
y = df_train['Target']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
numeric_features = ['BMI', 'MentHlth', 'PhysHlth']
```

```python
scaler = StandardScaler()
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_val[numeric_features] = scaler.transform(X_val[numeric_features])
```

```python
clf = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_val)
```

```python
balanced_accuracy = balanced_accuracy_score(y_val, y_pred)
print(f'Balanced Accuracy: {balanced_accuracy}')
```

```python
X_test = df_test.drop(['Id'], axis=1)
X_test[numeric_features] = scaler.transform(X_test[numeric_features])
```

```python
test_predictions = clf.predict(X_test)
```

```python
submission = pd.DataFrame({'Id': df_test['Id'], 'Target': test_predictions})
submission.to_csv('submission.csv', index=False)
```
